```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
URL = "https://simple.wikipedia.org/wiki/List_of_U.S._state_capitals"
```

```python
response= requests.get(URL)
```

```python
response.status_code
```

```python
print(response.text[:300])
```

```python
raw_HTML = BeautifulSoup(response.text,'html.parser')
```

```python
raw_HTML.find('table',class_='wikitable sortable')
```

```python
table = raw_HTML.find('table')
```

```python
tabledata= table.find_all('tr')[2:]
```

```python
state_info = {'State':[],
             'Abbreviation':[],
             'Statehood':[],
             'Capital':[],
              'Capital_Since':[],
              'Area':[],
              'City_population':[],
              'Metro_population':[],
              'State_rank':[],
              'US_rank':[],
              'Notes':[],}
             
```

```python
state_info.values()
```

```python
for index,item in enumerate(state_info.values()):
    print(index,item)
```

```python
for raw in tabledata:
    raw_data = raw.find_all('td')
    
    for index, item in enumerate(state_info.values()):
        
        data = raw_data[index].text
        
        if (data=="\n") or (data ==''):
            item.append(np.nan)
            
        else :
            item.append(data)
```
